## Data Cleaning & Preprocessing

```python
# Renaming the columns of the dataframe to make them more readable
df = df.rename(columns={
    'Invoice': 'InvoiceNo',
    'Price': 'UnitPrice',
    'Customer ID': 'CustomerID',
})
```

```python
# Perform any necessary data cleaning or handling of missing values.
df.isnull().sum()
```

```python
# Drop rows with missing CustomerID or Description
df.dropna(subset=['CustomerID', 'Description'], inplace=True)
```

```python
# Displaying the dataframe
df
```

```python
# Removing the duplicate rows from the dataframe
df.drop_duplicates(inplace=True)
```

```python
# Displaying the dataframe
df
```

```python
# Handle canceled orders (where InvoiceNo starts with 'C')
df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]
```

```python
# Handle Outliers (cap extreme values) in Quantity and UnitPrice
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)].copy()  # <- copy() prevents SettingWithCopyWarning
```

```python
# Convert InvoiceDate to proper datetime format
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

## Exploratory Data Analysis

### 1. General Overview:

```python
# General Overview: No. of unique products, No. of total transactions, No. of unique customers, No. of Countries
print("Unique Products:", df['StockCode'].nunique())
print("Total Transactions:", df['InvoiceNo'].nunique())
print("Unique Customers:", df['CustomerID'].nunique())
print("Countries:", df['Country'].nunique())
```

### 2. Product Analysis:

```python
# Top 10 Selling Products by Quantity
top_quantity = df.groupby('Description')['Quantity'].sum().nlargest(10)
top_quantity.plot(kind='barh', title='Top 10 Products by Quantity', figsize=(10,5))
plt.xlabel('Quantity Sold')
plt.show()
```

```python
# Top 10 Revenue-Generating Products
df['Revenue'] = df['Quantity'] * df['UnitPrice']
top_revenue = df.groupby('Description')['Revenue'].sum().nlargest(10)
top_revenue.plot(kind='barh', title='Top 10 Products by Revenue', color='orange', figsize=(10,5))
plt.xlabel('Total Revenue')
plt.show()
```

```python
# Products with Zero or Negative Prices
invalid_prices = df[df['UnitPrice'] <= 0]
invalid_prices[['StockCode', 'Description', 'UnitPrice']].drop_duplicates()
```

### 3. Customer Analysis:

```python
# Most products bought by customers
top_customers = df.groupby('CustomerID')['Quantity'].sum().nlargest(10)
top_customers.plot(kind='bar', title='Top Customers by Quantity', figsize=(10,5))
plt.ylabel('Quantity Purchased')
plt.show()
```

```python
# Distribution of Purchases per Customer
purchase_dist = df.groupby('CustomerID')['InvoiceNo'].nunique()
sns.histplot(purchase_dist, bins=30, kde=True)
plt.title("Distribution of Purchases per Customer")
plt.xlabel("Number of Purchases")
plt.show()
```

### 4. Time Series & Seasonal Trends:

```python
# Monthly Sales Trend

# Convert Period to string for better plotting
df['Month'] = df['InvoiceDate'].dt.to_period('M').astype(str)

# Group and plot
monthly_sales = df.groupby('Month')['Revenue'].sum()

plt.figure(figsize=(12, 5))
monthly_sales.plot(marker='o')
plt.title('Monthly Revenue Trend')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.xticks(rotation=0)  # Set rotation to 0 (flat)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Daily patterns of transactions

# Group by date
df['Date'] = df['InvoiceDate'].dt.date
daily_revenue = df.groupby('Date')['Revenue'].sum()

# Plot
plt.figure(figsize=(14,5))
daily_revenue.plot()
plt.title("Daily Revenue Trend")
plt.xlabel("Date")
plt.ylabel("Revenue")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Weekly patterns of transactions

# Add Day of Week column (0 = Monday, 6 = Sunday)
df['DayOfWeek'] = df['InvoiceDate'].dt.dayofweek

# Map to actual day names for better readability
dow_map = {0:'Mon', 1:'Tue', 2:'Wed', 3:'Thu', 4:'Fri', 5:'Sat', 6:'Sun'}
df['DayName'] = df['DayOfWeek'].map(dow_map)

# Group by Day Name
weekly_revenue = df.groupby('DayName')['Revenue'].mean().reindex(['Mon','Tue','Wed','Thu','Fri','Sat','Sun'])

# Plot
plt.figure(figsize=(8,5))
sns.barplot(x = weekly_revenue.index,
            y = weekly_revenue.values,
            hue = weekly_revenue.index,
            palette = 'Set2',
            legend = False)
plt.title("Average Revenue by Day of Week")
plt.xlabel("Day")
plt.ylabel("Average Revenue")
plt.tight_layout()
plt.show()
```

```python
# Identify peak sales months and slow seasons
df['Month'] = df['InvoiceDate'].dt.to_period('M').astype(str)

# Group by Month
monthly_revenue = df.groupby('Month')['Revenue'].sum()

# Identify Peak & Low months
peak_month = monthly_revenue.idxmax()
low_month = monthly_revenue.idxmin()

print(f"ðŸ“ˆ Peak Sales Month: {peak_month} with Revenue: {monthly_revenue.max():,.2f}")
print(f"ðŸ“‰ Lowest Sales Month: {low_month} with Revenue: {monthly_revenue.min():,.2f}")
```

```python
# Identify peak sales months and slow seasons
plt.figure(figsize=(12,5))
monthly_revenue.plot(marker='o', color='steelblue')
plt.axvline(x=peak_month, color='green', linestyle='--', label=f'Peak Month: {peak_month}')
plt.axvline(x=low_month, color='red', linestyle='--', label=f'Slow Month: {low_month}')
plt.title("Monthly Revenue Trend with Peaks Highlighted")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

### 5. Country Analysis:

```python
# Countries have the highest number of orders
country_orders = df.groupby('Country')['InvoiceNo'].nunique().sort_values(ascending=False).head(10)
country_orders.plot(kind='bar', title='Top Countries by Number of Orders', figsize=(10,5))
plt.ylabel("Number of Orders")
plt.show()
```

### 5. RFM (Recency, Frequency, Monetary) Segmentation:

```python
# Reference Date = Last date + 1
snapshot_date = df['InvoiceDate'].max() + pd.Timedelta(days=1)

rfm = df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (snapshot_date - x.max()).days,
    'InvoiceNo': 'nunique',
    'Revenue': 'sum'
})

rfm.columns = ['Recency', 'Frequency', 'Monetary']

# Assign RFM Scores safely
rfm['R'] = pd.qcut(rfm['Recency'], 4, labels=[4, 3, 2, 1])  # How recently a customer made a purchase (recent = higher score)
# Use cut (fixed bins) for skewed Frequency and Monetary
rfm['F'] = pd.cut(rfm['Frequency'], bins=4, labels=[1, 2, 3, 4]) # How often they purchased
rfm['M'] = pd.cut(rfm['Monetary'], bins=4, labels=[1, 2, 3, 4]) # How much they spent in total

# Combine into RFM Score
rfm['RFM_Score'] = rfm['R'].astype(str) + rfm['F'].astype(str) + rfm['M'].astype(str)
```

```python
# Explore the dataset by visualizing the relationships between different features.
sns.scatterplot(data=rfm, x='Recency', y='Monetary', hue='Frequency', palette='viridis')
plt.title("RFM Segmentation")
plt.show()
```
